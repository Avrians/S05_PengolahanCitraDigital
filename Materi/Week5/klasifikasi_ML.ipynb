```python
data = []
label = []
paths = []

for r, d, f in os.walk("/content/gdrive/MyDrive/Image Processing [PHB]/Image Classification/Latihan/data2/6 Cuaca/All/Clear"):
    for file in f:
        if '.jpg' in file:
            paths.append(os.path.join(r, file))

for path in paths:
    img = Image.open(path)
    x = img.resize((200, 200)) #seluruh langkah preprocessing dilakukan disini
    x = np.array(x)
    if(x.shape == (200, 200,3)):
        data.append(np.array(x).flatten()) #perbedaan deep learning tidak menggunakan flatten
        label.append(enc.transform([[0]]).toarray())

print (len(data))
print (len(label))
print (label)
```

```python
paths = []
for r, d, f in os.walk("/content/gdrive/MyDrive/Image Processing [PHB]/Image Classification/Latihan/data2/6 Cuaca/All/Haze"):
    for file in f:
        if '.png' in file:
          paths.append(os.path.join(r, file))

for path in paths:
    img = Image.open(path)
    x = img.resize((200,200)) #seluruh preprocessing dilakukan di sini
    x = np.array(x)
    if(x.shape == (200,200,3)):
      data.append(np.array(x).flatten())
      label.append(enc.transform([[1]]).toarray())

print (len(data))
print (len(label))
print (label)
```
