```python
import numpy as np

a = gmb1/255
c1 = 0.5
c2 = 2
c3 = 0.25
c4 = 1.5

log1 = c1*np.log(1 + (a))
log2 = c2*np.log(1 + (a))
log3 = c3*np.log(1 + (a))
log4 = c4*np.log(1 + (a))

plt.figure(figsize=(15,10))
plt.subplot(1,5,1); plt.imshow(gmb1)
plt.title('Original'); plt.axis("off")
plt.subplot(1,5,3); plt.imshow(log1)
plt.title('C = 0.5'); plt.axis("off")
plt.subplot(1,5,5); plt.imshow(log2)
plt.title('C = 2'); plt.axis("off")
plt.subplot(1,5,2); plt.imshow(log3)
plt.title('C = 0.25'); plt.axis("off")
plt.subplot(1,5,4); plt.imshow(log4)
plt.title('C = 1.5'); plt.axis("off")
plt.show()
```

```python
import numpy as np

a = gmb1/255
c = 1
p1 = 0.5
p2 = 2
p3 = 0.25
p4 = 1.5

gama1 = c*(a**p1)
gama2 = c*(a**p2)
gama3 = c*(a**p3)
gama4 = c*(a**p4)

plt.figure(figsize=(15,10))
plt.subplot(1,5,1); plt.imshow(gmb1)
plt.title('Original'); plt.axis("off")
plt.subplot(1,5,4); plt.imshow(gama1)
plt.title('P = 0.5'); plt.axis("off")
plt.subplot(1,5,2); plt.imshow(gama2)
plt.title('P = 2'); plt.axis("off")
plt.subplot(1,5,5); plt.imshow(gama3)
plt.title('P = 0.25'); plt.axis("off")
plt.subplot(1,5,3); plt.imshow(gama4)
plt.title('P = 1.5'); plt.axis("off")
plt.show()
```

```python
from skimage import filters, morphology, util

gray1 = color.rgb2gray(io.imread('dataset/rontgen.jpg'))
avg1 = filters.rank.mean(gray1,selem=morphology.square(3))
Gsn1 = util.img_as_ubyte(filters.gaussian(gray1, sigma=1))

plt.figure(figsize=(15,10))
plt.subplot(1,3,1);plt.imshow(gmb1,cmap='gray',vmin=0,vmax=255)
plt.title('Original'); plt.axis("off")
plt.subplot(1,3,2);plt.imshow(avg1,cmap='gray',vmin=0,vmax=255)
plt.title('Average Filter'); plt.axis("off")
plt.subplot(1,3,3);plt.imshow(Gsn1,cmap='gray',vmin=0,vmax=255)
plt.title('Gaussian Filter'); plt.axis("off")
plt.show()
```

```python
mawar = io.imread('dataset/noise1.jpg')
mawarGray = color.rgb2gray(mawar)
MdnFil=filters.rank.median(mawarGray,selem=morphology.square(3))
MdnAvg = filters.rank.mean(mawarGray,selem=morphology.square(3))

plt.figure(figsize=(15,10))
plt.subplot(1,3,1);plt.imshow(mawar,cmap='gray',vmin=0,vmax=255)
plt.title('Original'); plt.axis("off")
plt.subplot(1,3,2);plt.imshow(MdnFil,cmap='gray',vmin=0,vmax=255)
plt.title('Average Filter'); plt.axis("off")
plt.subplot(1,3,3);plt.imshow(MdnAvg,cmap='gray',vmin=0,vmax=255)
plt.title('Gaussian Filter'); plt.axis("off")
plt.show()
```

```python
latihan_noise = io.imread('dataset/latihan_noise.jpg')
MdnFil2=filters.rank.median(latihan_noise,selem=morphology.square(3))
MdnAvg2 = filters.rank.mean(latihan_noise,selem=morphology.square(3))

plt.figure(figsize=(25,10))
plt.subplot(1,3,1);plt.imshow(latihan_noise,cmap='gray',vmin=0,vmax=255)
plt.title('Original'); plt.axis("off")
plt.subplot(1,3,2);plt.imshow(MdnFil2,cmap='gray',vmin=0,vmax=255)
plt.title('Average Filter'); plt.axis("off")
plt.subplot(1,3,3);plt.imshow(MdnAvg2,cmap='gray',vmin=0,vmax=255)
plt.title('Gaussian Filter'); plt.axis("off")
plt.show()
```
