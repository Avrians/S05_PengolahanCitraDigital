```python
from skimage import color, io
from skimage.transform import rescale
import matplotlib.pyplot as plt
```

```python
gmb1 = io.imread('dataset/gelap1.jpg')
plt.imshow(gmb1)
plt.show()
```

```python
gmb2 = io.imread('dataset/gelap2.jpg')
plt.imshow(gmb2, cmap='gray')
plt.show()
```

```python
plt.subplot(1, 3, 1); plt.imshow(gmb1) #bisa menggunakan ; atau enter
plt.subplot(1, 3, 2)
plt.imshow(gmb2)
plt.show()
```

```python
import numpy as np

a = gmb1/255
b = gmb2/255

c1 = 5
c2 = 15
c3 = 1
c4 = 10

log1 = c1*np.log(1 + (a))
log2 = c2*np.log(1 + (a))
log3 = c3*np.log(1 + (a))
log4 = c4*np.log(1 + (a))

log11 = c1*np.log(1 + (b))
log12 = c2*np.log(1 + (b))
log13 = c3*np.log(1 + (b))
log14 = c4*np.log(1 + (b))

plt.figure(figsize=(25,20))
plt.subplot(1,5,1); plt.imshow(gmb1)
plt.title('Original'); plt.axis("off")
plt.subplot(1,5,3); plt.imshow(log1)
plt.title('C = 5'); plt.axis("off")
plt.subplot(1,5,5); plt.imshow(log2)
plt.title('C = 15'); plt.axis("off")
plt.subplot(1,5,2); plt.imshow(log3)
plt.title('C = 1'); plt.axis("off")
plt.subplot(1,5,4); plt.imshow(log4)
plt.title('C = 10'); plt.axis("off")
plt.show()

plt.figure(figsize=(25,20))
plt.subplot(1,5,1); plt.imshow(gmb2)
plt.title('Original'); plt.axis("off")
plt.subplot(1,5,3); plt.imshow(log11)
plt.title('C = 5'); plt.axis("off")
plt.subplot(1,5,5); plt.imshow(log12)
plt.title('C = 15'); plt.axis("off")
plt.subplot(1,5,2); plt.imshow(log13)
plt.title('C = 1'); plt.axis("off")
plt.subplot(1,5,4); plt.imshow(log14)
plt.title('C = 10'); plt.axis("off")
plt.show()
```
